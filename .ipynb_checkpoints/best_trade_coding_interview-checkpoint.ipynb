Say you have an array for which the ith element is the price of a given stock on day i.

If you were only permitted to complete at most one transaction (i.e., buy one and sell one share of the stock), design an algorithm to find the maximum profit.

Note that you cannot sell a stock before you buy one.

```python
def maxProfit(prices):
''' prices: 'List[int]' 
    return: [int]'''
    n = len(prices)
    if n == 0 or n == 1:
        return 0

    cur_min, cur_max, cur_diff = float('inf'), 0, 0
    best_min, best_max, best_diff = float('inf'), 0, 0
    cur_min_idx, best_min_idx = 0, 0
#     cur_max_idx, best_max_idx = 0, 0   # these aren't needed for this problem

    for i in range(0, n):
        p = prices[i]

        if p < cur_min:
            cur_min = p
            cur_min_idx = i
            # reset diff & max:
            cur_diff = 0
            cur_max = 0

        if p > cur_max:
            cur_max = p
            cur_diff = cur_max - cur_min
            cur_max_idx = i

        if cur_diff > best_diff:
            best_min = cur_min
#             best_max = cur_max   # not needed
            best_diff = cur_diff
            best_min_idx = cur_min_idx
            best_max_idx = cur_max_idx

    return best_diff
```

```python
# can shorten even further since don't need indices
def maxProfit(prices):
    n = len(prices)
    if n == 0 or n == 1:
        return 0

    cur_min, cur_max, cur_diff = prices[0], 0, 0
    best_min, best_diff = prices[0], 0

    for i in range(1, n):
        p = prices[i]

        if p < cur_min:
            cur_min = p
            # reset diff & max:
            cur_diff = 0
            cur_max = p

        if p > cur_max:
            cur_max = p
            cur_diff = cur_max - cur_min

        if cur_diff > best_diff:
            best_diff = cur_diff

    return best_diff
```
